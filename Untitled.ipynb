# Implementing with sklearn library


```python
from sklearn.cluster import KMeans
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
from scipy.spatial.distance import cdist
```

```python
img = cv2.imread('kmean.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

ret,img = cv2.threshold(img,172,255,cv2.THRESH_BINARY_INV)
```

```python

plt.imshow(img,cmap = 'gray')
```

```python
kernel = np.ones((3,2),np.uint8)
img = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(img,cmap = 'gray')
```

```python
img1 = img.copy()
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

X = []
for contour in contours:
    x,y,w,h = cv2.boundingRect(contour)
    X.append([x,y])
    cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255,0), 2)

X = np.reshape(X,(-1,2))
```

```python
X.shape
```

```python
model = KMeans(n_clusters=6, random_state=0)
```

```python
model.fit(X)
```

```python
labels = model.labels_
```

```python
plt.plot(X[labels == 0,0],X[labels == 0,1],'b^')
plt.plot(X[labels == 1,0],X[labels == 1,1],'r^')
plt.plot(X[labels == 2,0],X[labels == 2,1],'bo')
plt.plot(X[labels == 3,0],X[labels == 3,1],'go')
plt.plot(X[labels == 4,0],X[labels == 4,1],'g^')
plt.plot(X[labels == 5,0],X[labels == 5,1],'ro')
```

# Implementing with Numpy 

```python
def kmeans_init_centroids(X,k):
    index = np.random.permutation(X.shape[0])[:k]
    return X[index]
```

```python
def kmean_assign_labels(X,centroid):
    dis = cdist(X,centroid)
    index = np.argmin(dis,axis = 1)
    return index
```

```python
def kmeans_update_centroid(X,labels):
    k = len(np.unique(labels))
    centroids = np.zeros((k,X.shape[1]))
    for i in range(k):
        Xi = X[labels == i,:]
        centroids[i] = np.mean(Xi,axis = 0)
    return centroids
        
    
    
```

```python
def has_converged(centroids, new_centroids):
# return True if two sets of centroids are the same
    return (set([tuple(a) for a in centroids]) == set([tuple(a) for a in new_centroids]))
```

```python
centroids = []
centroids.append(kmeans_init_centroids(X,6))
```

```python
while(True):
    labels = kmean_assign_labels(X,centroids[-1])
    new_centroids = kmeans_update_centroid(X,labels)
    if has_converged(centroids[-1],new_centroids):
        break
    centroids.append(new_centroids)
```

```python
labels = kmean_assign_labels(X,new_centroids)
```

```python
plt.plot(X[labels == 0,0],X[labels == 0,1],'b^')
plt.plot(X[labels == 1,0],X[labels == 1,1],'r^')
plt.plot(X[labels == 2,0],X[labels == 2,1],'bo')
plt.plot(X[labels == 3,0],X[labels == 3,1],'go')
plt.plot(X[labels == 4,0],X[labels == 4,1],'g^')
plt.plot(X[labels == 5,0],X[labels == 5,1],'ro')
```
